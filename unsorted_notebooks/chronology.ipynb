```python
import pandas as pd

moonshot = pd.read_pickle('moonshot_submissions.p')
```

```python
moonshot[['ORDER_DATE', 'Order Date', 'Shipment Date', 'SHIPMENT_DATE']]
```

```python
import datetime

dates = moonshot.loc[moonshot.ORDER_DATE.astype(bool)].ORDER_DATE.apply(datetime.date.fromisoformat)
```

```python
import plotly.express as px
```

```python
fig = px.histogram(dates,
                   title='Order date, monthly bin')
fig.update_layout(showlegend=False)
fig
```

```python
xdates = moonshot.loc[moonshot.ORDER_DATE.astype(bool) &
                      (moonshot['Structure ID'] != '')]\
                  .ORDER_DATE.apply(datetime.date.fromisoformat)
fig = px.histogram(xdates,
                   title='Order date, xstallised, monthly bin')
fig.update_layout(showlegend=False)
fig
```
